```python
import json
import matplotlib.pyplot as plt

# Load custom logs from JSON file
log_file = 'custom_logs_epoch_testseq_1.json'
with open(log_file, 'r') as file:
    custom_logs = json.load(file)

# Extract steps and test_loss
steps = []
losses = []

for step_log in custom_logs['steps']:
    steps.append(step_log['step'])        # Extract step number
    losses.append(step_log['test_loss'])  # Extract test_loss

# Plotting loss vs step
plt.figure(figsize=(10, 6))
plt.plot(steps, losses, marker='o', linestyle='-', color='b', label='Test Loss')
plt.title('Loss vs Step Number')
plt.xlabel('Step Number')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Load custom logs from JSON file
log_file = 'custom_logs_epoch_testseq_1.json'
with open(log_file, 'r') as file:
    custom_logs = json.load(file)

# Extract steps, queue delays, and test_losses
steps = []
queue_delays = []  # List to store queue delays
losses = []  # List to store test losses

for step_log in custom_logs['steps']:
    steps.append(step_log['step'])  # Extract step number
    queue_delay = step_log['states'][0][0][3]  # Extract queue delay
    queue_delays.append(queue_delay)  # Store queue delay
    losses.append(step_log['test_loss'])  # Extract test loss

# Plotting queue delay vs step
plt.figure(figsize=(10, 6))
plt.plot(steps, queue_delays, marker='o', linestyle='-', color='r', label='Queue Delay')
plt.title('Queue Delay vs Step Number')
plt.xlabel('Step Number')
plt.ylabel('Queue Delay')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Optional: If you still want to plot test loss vs step, you can do it as well:
# Plotting loss vs step
plt.figure(figsize=(10, 6))
plt.plot(steps, losses, marker='o', linestyle='-', color='b', label='Test Loss')
plt.title('Loss vs Step Number')
plt.xlabel('Step Number')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
